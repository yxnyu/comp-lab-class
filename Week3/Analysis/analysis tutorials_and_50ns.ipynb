```python
data=pd.read_csv('/home/yx2892/pressure.xvg',header=0,names=['Time(ps)','Pressure(bar)'],skiprows=24,sep='\s+')
display( data.head())
data[ '10-ps-steps_rolling_avg' ] = data['Pressure(bar)'].rolling( 10).mean()
#data['Time']=data['Time']/1000

    
plt.plot(data['Time(ps)'],data['Pressure(bar)'],label='Pressure(bar)')
plt.plot(data['Time(ps)'],data[ '10-ps-steps_rolling_avg' ],label='ten_avg_running(bar)')
plt.ylabel('Pressure(bar)')
plt.xlabel('Time (ps)')
plt.xlim(0,100)
#plt.ylabel('Pressure-(bar)')
plt.title('Pressure')
plt.legend()
plt.show()    
    

```
