```python
# this notebook is for plotting
# the end to end analysis
import numpy as np
import matplotlib.pyplot as plt
#load mdtraj
import mdtraj as md
```

```python
trr_file='adp_exchange4temps.trr'
gro_file='adp_exchange4temps.gro'
traj=md.load(trr_file,top=gro_file)
a,phi = md.compute_phi(traj)
b,psi = md.compute_psi(traj)
```

```python
print(phi)

#print(psi)
```

```python
time=traj.time/1000
print(time)
```

```python
#plot distance vs time
plt.plot(traj.time/1000,phi,label='phi')
plt.xlabel('Time(ns)')
plt.ylabel('angel')
plt.title('phi')
plt.show()
```

```python
plt.plot(traj.time/1000,psi,label='psi')
plt.xlabel('Time(ns)')
plt.ylabel('Angel')
plt.title('psi')
plt.show()
```

```python
#Make a 2d histogram of the data
psi_,phi_=[],[]
for i in psi:
    psi_.append(i[0])
for i in phi:
    phi_.append(i[0])

H,xedges, yedges =np.histogram2d(phi_, psi_, bins=10)

H =-np.log(H)
H=H-np.min(H)

plt.imshow(H,interpolation='nearest',origin='lower',extent=[xedges[0],xedges[-1],yedges[0],yedges[-1]])
plt.xlabel('phi(rad)')
plt.ylabel('phi(rad)')
plt.colorbar()
plt.show()
```
