```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('replica_temp.xvg',names=['Time','Index1','Index2','Index3','Index4'],sep='\s+')
```

```python
time=[]
r1=[]
r2=[]
r3=[]
r4=[]
for i in range (50000):
    if(i%5000==0):
        time.append(data['Time'][i]/1000)
        r1.append(data['Index1'][i]*50+300)
        r2.append(data['Index2'][i]*50+300)
        r3.append(data['Index3'][i]*50+300)
        r4.append(data['Index4'][i]*50+300)
    
    
```

```python
fig,ax=plt.subplots()
ax.plot(time,r1,label='T=300',color='blue')
ax.plot(time,r2,label='T=400',color='green')
ax.plot(time,r3,label='T=450',color='red')
ax.plot(time,r4,label='T=600',color='black')
ax.legend(loc=(1.1,0.5))
ax.set_xlabel("Time(ns)")
ax.set_ylabel("Temperature(K)")
ax.set_title("Replica exchange")
```
