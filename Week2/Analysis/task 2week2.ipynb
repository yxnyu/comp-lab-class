```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data=pd.read_csv('1hz3_T310.run.25000000.energy.xvg',header=0,names=['Time','Potential','Kinetic En.','Total Energy','Temperature','Pressure','Volume'],skiprows=29,sep='\s+')
print(data)

#6 plots 
data.plot(x='Time',y='Potential')
data.plot(x='Time',y='Kinetic En.')
data.plot(x='Time',y='Total Energy')
data.plot(x='Time',y='Temperature')
data.plot(x='Time',y='Pressure')
data.plot(x='Time',y='Volume')

#









```

```python
x = data.iloc[1:,0]
print(x)
y1= data.iloc[1:,1]
y2= data.iloc[1:,4]
 
fig,ax1 = plt.subplots()
ax2 = ax1.twinx()           # 做镜像处理
ax1.plot(x,y1,'g-')
ax2.plot(x,y2,'b--')
 
ax1.set_xlabel('time')    #设置x轴标题
ax1.set_ylabel('Potential',color = 'g')   #设置Y1轴标题
ax2.set_ylabel('Temperature',color = 'b')   #设置Y2轴标题

```
